***✲ This analysis examines how book sales vary by the original language of the books in the dataset.***

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.barplot(x='Original_language', y='Approximate_sales', data=df, palette='viridis')
plt.xticks(rotation=45)
plt.title("Sales By Original Languages")
plt.xlabel("Original Language")
plt.ylabel("Sales")

plt.subplot(1,2,2)
sns.swarmplot(x=df.Original_language,y=df.Approximate_sales,palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Original Language")
plt.ylabel("Sales")

plt.show()
```

***✲ In this step, a new column, Age_of_books, is created to calculate the age of each book based on its First_published year.
The result is a new column representing the age of each book, which helps in analyzing the relationship between a book's age and other factors, such as sales or genre trends.***

```python
current_year=2024

def age_Calculator(x):
    return current_year - int(x)
```

```python
df["Age_of_books"] = df["First_published"].transform(age_Calculator)
```

```python
df.Age_of_books
```

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(6, 4),facecolor='lavender') 
sns.histplot(
    x=df.Age_of_books,
    y=df.Approximate_sales,
    cmap="coolwarm",  
    cbar=True
)
plt.xlabel("Age of Books", color="darkblue")
plt.ylabel("Sales", color="darkblue")

plt.show()
```

***✲ A new column, Sales_category, is added to the dataset to classify books based on their sales figures.The Sales_category column categorizes each book's sales into predefined groups, allowing for easier analysis of sales trends and comparisons across different sales ranges.***

```python
def sales_label(x):
    if int(x) >= 120:
        return 'High'
    if int(x) >=50 and int(x)<120:
        return 'Medium'
    if int(x) >= 0 and int(x) <50:
        return 'Low'
```

```python
df["Sales_category"]=df["Approximate_sales"].transform(sales_label)
```

```python
df.Sales_category.value_counts()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10), constrained_layout=True)


sns.boxplot(
    x='Sales_category',
    y='Approximate_sales',
    data=df,
    ax=axes[0][0],
    palette='pastel'
)
axes[0][0].set_title('Sales Category vs Approximate Sales',fontsize=16,color="gold")
axes[0][0].set_xlabel('Sales Category')
axes[0][0].set_ylabel('Approximate Sales')


df['Sales_category'].value_counts().plot(
    kind='barh',
    color=sns.color_palette('viridis'),
    ax=axes[0][1]
)
axes[0][1].set_xlabel('Count')
axes[0][1].set_ylabel('Sales Category')
axes[0][1].set_title('Sales Category Counts',fontsize=16,color="gold")


sns.swarmplot(
    x='Sales_category',
    data=df,
    ax=axes[1][0],
    palette='pastel'
)
axes[1][0].set_xlabel('Sales Category')
axes[1][0].set_ylabel('Count')
axes[1][0].set_title('Distribution of Sales Categories',fontsize=16,color="gold")


df['Sales_category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=sns.color_palette('Set2'),
    startangle=90,
    ax=axes[1][1],
    fontsize=12
)
axes[1][1].set_ylabel('')
axes[1][1].set_title('Percentage of Sales Categories',fontsize=16,color="gold")

plt.tight_layout()
plt.show()
```
